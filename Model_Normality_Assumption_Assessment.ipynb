## Loading Data

```r
df = read.csv('./data/alcoholtobacco.csv', sep='\t', fileEncoding='UTF-16LE')
head(df)
```

## Fitting Model

```r
model = lm(Alcohol~Tobacco, data=df)
model.summ = summary(model)
model.summ
```

```r
# the residuals
res = model.summ$residuals
t(res)
```

## QQ-plot

```r
options(repr.plot.height=5)
qqnorm(res, pch = 19, col='blue')
qqline(res, col = 'red', lwd = 2, lty=3)
```

## Histogram

```r
hist(res, xlab='Residuals', main='')
```
