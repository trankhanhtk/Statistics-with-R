# Monotone Data

## Non-leanear Monotone

```r
if (! exists('options_org')) options_org = options()

x=c(1, 2, 5, 6, 9, 10, 11, 13)
y=c(2, 3, 4, 5, 7, 9, 11, 12)
df = data.frame(x = x, y = y)
df = df[order(df$x),]

matrix(c('X', x, 'Y', y), nrow=2, byrow=T)
matrix(c('RANK(X)', rank(x), 'RANK(Y)', rank(y)), nrow=2, byrow=T)

options(repr.plot.width=7, repr.plot.height=5)
par(mfrow=c(1,1), bg='gray90')

plot(rank(x), rank(y), xlab='', ylab='', pch=19, xlim=c(1, 13), ylim=c(1, 12), xaxt='n', yaxt='n', col='red')
abline(lsfit(rank(x), rank(y), intercept=T), col='red', lty=2)
points(df$x, df$y, pch=18, col='blue', type='b', lty=2)

legend(1, 12, legend = c("Rank X/Rank Y", "X/Y"), col = c('red', 'blue'), lty = 2:2, y.intersp=1.5)

axis(1, at=seq(1, 13, by=1))
axis(2, at=seq(1, 12, by=1), las=1)
```

## Linear Monotone

```r
x = c (4, 3, -.5, 2.5, 5, 2)
y = x*2 + 1
df = data.frame(x = x, y = y)
df = df[order(df$x),]

plot(rank(x), rank(y), pch=19, col='red', xaxt='n', yaxt='n', xlab='', ylab='', xlim=c(-1, 6), ylim=c(-1, 10))
points(df$x, df$y, pch=18, col=rgb(0, 0, 1))
abline(lsfit(rank(x), rank(y)), col='red', lty=2)
abline(lsfit(x, y), col='blue', lty=2)

legend(-1, 10, legend = c("Rank X/Rank Y", "X/Y"), col = c('red', 'blue'), lty = 2:2, y.intersp=1.5)

axis(1, at=seq(-1, 6, by=1))
axis(2, at=seq(-1, 10, by=1), las=2)
```

### Remarks
Monotone data have their ranks lie on a line.

## Ties

```r
if (! exists('options_org')) options_org = options()

x=c(1, 3, 5, 6, 9, 10, 11, 13)
y=c(2, 3, 4, 5, 7, 9, 9, 12)
df = data.frame(x = x, y = y)
df = df[order(df$x),]

matrix(c('X', x, 'Y', y), nrow=2, byrow=T)
matrix(c('RANK(X)', rank(x), 'RANK(Y)', rank(y)), nrow=2, byrow=T)

par(mfrow=c(1,1), bg='gray90')
plot(rank(x), rank(y), xlab='', ylab='', pch=19, xlim=c(1, 13), ylim=c(1, 12), xaxt='n', yaxt='n', col='red')
points(df$x, df$y, pch=18, col='blue', type='b', lty=2)

text(6.25, 8.5, 'rank(y)=6.5', col='red')
arrows(x0=6.25, y0=8, x1=6, y1=6.7, col='red', length=.1)
arrows(x0=6.25, y0=8, x1=6.9, y1=6.7, col='red', length=.1)

abline(lsfit(rank(x), rank(y), intercept=T), col='red', lty=2)
abline(v=7, lty=2, col='gray75')

legend(1, 12, legend = c("Rank X/Rank Y", "X/Y"), col = c('red', 'blue'), lty = 2:2, y.intersp=1.5)

axis(1, at=seq(1, 13, by=1))
axis(2, at=seq(1, 12, by=1), las=1)
```
