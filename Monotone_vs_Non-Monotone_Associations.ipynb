## Monotone

```r
if (! exists('options_org')) options_org = options()

x=c(1, 2, 5, 6, 9, 10, 11, 13)
y=c(2, 3, 4, 5, 7, 9, 11, 12)
len = length(x)

matrix(c('X', x, 'Y', y), nrow=2, byrow=T)

matrix(c('RANK(X)', rank(x), 'RANK(Y)', rank(y)), nrow=2, byrow=T)


options(repr.plot.width=7, repr.plot.height=5)
par(mfrow=c(1,1), bg='gray90')
plot(rank(x), rank(y), xlab='rank(x) / x', ylab='rank(y) / y', pch=19, xlim=c(1, 13), ylim=c(1, 12), xaxt='n', yaxt='n', col='red')
axis(1, at=seq(1, 13, by=1), las=2)
axis(2, at=seq(1, 12, by=1),las=1)

points(x, y, pch=19, col=rgb(0, 0, 1, .8), cex=.75)
abline(lsfit(rank(x), rank(y), intercept=T), col='orange', lty='dotted')
```

### Remarks
Monotone data have their ranks lie on a line.

## Linear Monotone

```r
x = c (4, 3, -1.3, 2.2, 5, 2)
y = x*1.2 + .5
plot(rank(x), rank(y), pch=19, col='red')
points(x, y, pch=20, col=rgb(0, 0, 1, .8))
abline(lsfit(rank(x), rank(y)), col='orange', lty='dotted')
abline(lsfit(x, y), col='blue', lty='dotted')
```

## Ties

```r
if (! exists('options_org')) options_org = options()

x=c(1, 3, 5, 6, 9, 10, 11, 13)
y=c(2, 3, 4, 5, 7, 9, 9, 12)
len = length(x)

matrix(c('X', x, 'Y', y), nrow=2, byrow=T)

matrix(c('RANK(X)', rank(x), 'RANK(Y)', rank(y)), nrow=2, byrow=T)


options(repr.plot.width=7, repr.plot.height=5)
par(mfrow=c(1,1), bg='gray90')
plot(rank(x), rank(y), xlab='rank(x) / x', ylab='rank(y) / y', pch=19, xlim=c(1, 13), ylim=c(1, 12), xaxt='n', yaxt='n', col='red')
axis(1, at=seq(1, 13, by=1), las=2)
axis(2, at=seq(1, 12, by=1),las=1)

points(x, y, pch=19, col=rgb(0, 0, 1, .8), cex=.75)
text(6.25, 8.5, 'rank(y)=6.5', col='red')
arrows(x0=6.25, y0=8, x1=6, y1=6.7, col='red', length=.1)
arrows(x0=6.25, y0=8, x1=6.9, y1=6.7, col='red', length=.1)
abline(lsfit(rank(x), rank(y), intercept=T), col='orange', lty='dotted')
abline(v=7, lty='dotted', col='gray75')
```
