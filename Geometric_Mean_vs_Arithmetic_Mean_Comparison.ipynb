## Motivating Problem
Consider the following question (it's one of the questions in a Coursera assignment):

> You simulate a lot of $lognormal(5,1)$ random variables and take their mean. What is this number closest to?

And there are two valid options for the question are $e^{\mu + .5}$ and $e^{\mu}$ where $\mu = 5$.

Why are the two options are valid? Recall that the geometric mean is always less than or equal to the arithmetic mean, or symbolically:

\begin{align}
\left( \prod_{i=1}^{n}x_i \right)^{1/n} \le \frac{\sum_{i=1}^{n}x_i}{n}.
\end{align}

Where the quantity of the left side of the inequality is the geometric mean or $e^{\mu}$ in this case and the right-side quantity is the arithmetic mean or the mean of the sampling data of the log-normal variable.

Let's verify each of those options by employing simulation. For this purpose, let the random variable $L_{X}\sim N(5,1)$, then $X = e^{L_{X}}$ is a log-normal RV.

## Sampling Data of $L_{X}$

```r
nosim = 1000
lx = rnorm(nosim, 5, 1)
```

```r
options(repr.plot.height=5)
hist(lx, probability=T, density=20, col='darkblue', xlab=expression(L[X]), main='')
```

```r
cat('mean(lx):', mean(lx))
head(lx)
tail(lx)
```

## The Corresponding Sampling Data of $X$

```r
x = exp(lx)
```

```r
hist(x, probability=T, col=rgb(0, 0, 1, .25), main='', breaks=150, xlab='X')
```

```r
cat('mean(X):', mean(x))
head(x)
tail(x)
```

## Comparing the Results

```r
cat('\nexp(5+.5):', exp(5+.5))
cat('\nexp(5):', exp(5))
```

## Conclusion
From the empirical results, it's reasonable to choose the option $e^{\mu + .5}$. Can we answer the question without the aid of computer?
