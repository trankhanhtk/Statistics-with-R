## QQ Plot of an Exponential RV vs.  a Uniform RV
Let generate data for the uniform random variable $X \sim exp(1)$ and plot it over the corresponding quantile of $Y \sim U(0, 1)$.

The points should closely follow the curve of the following function:

\begin{align}
f(x) = -ln(1-x)
\end{align}

### Generate Data

```r
x_gen_ordered = sort(rexp(99))
probs = seq(0, 1, length=100)[-100]
y_quantiles = qunif(probs)
```

### Check Lengths & Data

```r
cat('length x_gen_ordered:', length(x_gen_ordered))
head(x_gen_ordered)
tail(x_gen_ordered)

cat('length q_exp:', length(q_exp))
head(q_exp)
tail(q_exp)
```

## The Q-Q Plot

```r
options(repr.plot.height=5)
plot(y_quantiles, x_gen_ordered, pch=19, cex=.5, col='blue', xlab='Sample quantiles', ylab='Hypothetical U(0,1) quantiles')
## Add a curve of f(x) = -ln(1-x)
curve(-log(1-x), col='red', add=T)
```
