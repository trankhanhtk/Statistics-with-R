## Loading Data & Libraries

```r
library('tidyverse')
```

```r
df = read.csv('./data/alcoholarm.csv')
head(df)
```

## Fitting the Simple Linear Regression Model

```r
model <- lm(strength ~ alcohol, data=df)
model.sum <- summary(model)
model.sum
```

## Regression Plot

```r
if (!exists('opts')) opts = options()

options(repr.plot.height=5)
plot(df$alcohol, df$strength, pch=19, col='darkblue', xlab='Alcohol', ylab='Arm Strength')
abline(a = model.sum$coefficients[1], b = model.sum$coefficients[2], col='red', lwd=2)
```

## Fitted Values vs. Residuals Plot

```r
y_hat <- model.sum$coefficients[1] + model.sum$coefficients[2]*df$alcohol
res <- df$strength - y_hat
plot(y_hat, res, xlab='Fitted Values', ylab='Residuals', pch=19, col='darkblue')
abline(h=0, col='black', lty='dotted', lwd=2)
```

## Regressor vs. Residuals Plot

```r
plot(df$alcohol, res, xlab='Regressor', ylab='Residuals', pch=19, col='darkblue')
abline(h=0, col='black', lty='dotted', lwd=2)
```
