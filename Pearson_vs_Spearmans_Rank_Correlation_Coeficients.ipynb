# Comparison Between Pearson correlation coefficient and Spearman's Rank Correlation Coeficients

## Data Without Outliers

```r
dat = read.csv('./data/fuel_consumption.txt', header=T)
cat(sum(rank(dat$Fuel.Consumption) != dat$Ranks.of.Fuel.Consumption), '<- Should equal zero\n')
cat(sum(rank(dat$Equivalence.Ratio) != dat$Ranks.of.Equivalence.Ratio), '<- Should equal zero\n')
dat
```

### Pearson and Spearman's Rank Correlation Coefficients

```r
pcorr1 = round(cor(dat$Equivalence.Ratio, dat$Fuel.Consumption, method='pearson'), digits=2)
scorr1 = round(cor(dat$Equivalence.Ratio, dat$Fuel.Consumption, method='spearman'), digits=2)
pcorr2 = round(cor(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, method='pearson'), digits=2)
scorr2 = round(cor(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, method='spearman'), digits=2)
cat(sprintf('Pearson/Spearman on Data: %s/%s\n', pcorr1, scorr1))
cat(sprintf('Pearson/Spearman on Ranks of Data: %s/%s\n', pcorr2, scorr2))
```

### The Plots

```r
if ( ! exists('options_org'))
    options_org = options()

options(repr.plot.width=8, repr.plot.height=5)

par(mfrow=c(1, 2), bg='gray90', mgp=c(2.2, 0.8, 0))
plot(dat$Equivalence.Ratio, dat$Fuel.Consumption, xlab='Equivalence Ratio', ylab='Fuel Consumption', pch=20, col='darkblue', cex.axis=.7, main=sprintf('Pearson/Spearman (%s/%s)', pcorr1, scorr1))
plot(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, xlab='Ranks of Equivalence Ratio', ylab='Ranks of Fuel Consumption', pch=20, col='darkblue', cex.axis=.7, main=sprintf('Pearson/Spearman (%s/%s)', pcorr2, scorr2))
```

## Data with Outlier

```r
dat[20,] = c(95, 2.1, NA, NA)
dat$Ranks.of.Fuel.Consumption=rank(dat$Fuel.Consumption)
dat$Ranks.of.Equivalence.Ratio=rank(dat$Equivalence.Ratio)
dat
```

### Pearson and Spearman Correlation Coefficients

```r
pcorr1 = round(cor(dat$Equivalence.Ratio, dat$Fuel.Consumption, method='pearson'), digits=2)
scorr1 = round(cor(dat$Equivalence.Ratio, dat$Fuel.Consumption, method='spearman'), digits=2)
pcorr2 = round(cor(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, method='pearson'), digits=2)
scorr2 = round(cor(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, method='spearman'), digits=2)
cat(sprintf('Pearson/Spearman on Data: %s/%s\n', pcorr1, scorr1))
cat(sprintf('Pearson/Spearman on Ranks of Data: %s/%s\n', pcorr2, scorr2))
```

### The Plots

```r
par(mfrow=c(1, 2), bg='gray90', mgp=c(2.2, 0.8, 0))
plot(dat$Equivalence.Ratio, dat$Fuel.Consumption, xlab='Equivalence Ratio', ylab='Fuel Consumption', pch=20, col='darkblue', cex.axis=.7, main=sprintf('Pearson/Spearman (%s/%s)', pcorr1, scorr1))
plot(dat$Ranks.of.Equivalence.Ratio, dat$Ranks.of.Fuel.Consumption, xlab='Ranks of Equivalence Ratio', ylab='Ranks of Fuel Consumption', pch=20, col='darkblue', cex.axis=.7, main=sprintf('Pearson/Spearman (%s/%s)', pcorr2, scorr2))
```

## Key Point
The Pearson correlation coefficient is less resistant to outliers compared with the Spearman's rank correlation coefficient. More over, Pearson corr. coef. only measures the leanear association.
