## Weibull Density Plot

```r
library(ggplot2)

options_org <- options()
options(repr.plot.width=10, repr.plot.height=9)

x_lim = 10
y_lim = 2
ggp = ggplot(
    data.frame(x=c(0, x_lim)), aes(x)) +
    scale_x_continuous('X', breaks = seq(0, x_lim, length=11), limits=c(0, x_lim)) +
    scale_y_continuous('Density', breaks = seq(0, y_lim, length=11), limits=c(0, y_lim)) +

    stat_function(fun = dweibull, args = list(shape=1, scale=0.5), aes(colour = 'W(1, 0.5)') ) +
    stat_function(fun = dweibull, args = list(shape=1, scale=1), aes(colour = 'W(1, 1)') ) +
    stat_function(fun = dweibull, args = list(shape=1, scale=5), aes(colour = 'W(1, 5)'), linetype='dotted' ) +
    stat_function(fun = dweibull, args = list(shape=1.5, scale=1), aes(colour = 'W(1.5, 1)') ) +
    stat_function(fun = dweibull, args = list(shape=2, scale=2), aes(colour = 'W(2, 2)') ) +
    stat_function(fun = dweibull, args = list(shape=2, scale=5), aes(colour = 'W(2, 5)'), linetype='dashed' ) +
    stat_function(fun = dweibull, args = list(shape=5, scale=1), aes(colour = 'W(5, 1)') ) +
    stat_function(fun = dweibull, args = list(shape=5, scale=5), aes(colour = 'W(5, 5)'), linetype='twodash' ) +
    theme(
        legend.position = c(.95, .95),
        legend.justification = c("right", "top"),
        legend.box.just = "right",
        legend.margin = margin(6, 6, 6, 6)
    ) +
    scale_color_manual('Weibull Density Distribution Functions',  values=c('W(1, 0.5)'='blue', 'W(1, 1)'='violet', 'W(1, 5)'='black', 'W(1.5, 1)'='orange', 'W(2, 2)'='cyan', 'W(2, 5)'='red', 'W(5, 1)'='green', 'W(5, 5)'='#e28743'))
ggp
options(options_org)
```
