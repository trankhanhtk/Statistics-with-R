## PDF of the Exponential Distribution

```r
options_org = options()
```

```r
?dexp
```

```r
x = seq(0, 10, length=101)
rates = c(.1, .5, 1, 1.5)

options(repr.plot.width=8, repr.plot.height=6)

par(mfrow=c(2,2), bg='gray95')
layout(mat = matrix(c(1, 2, 3, 4), nrow = 2, ncol = 2))

for (i in rates) {
    dx = dexp(x)

    plot(x, dx, col='blue', type='l', xlab='X', ylab='Density')
}
```

```r
x = seq(0, 10, length=101)
dx = dexp(x, 1)
options(repr.plot.width=7, repr.plot.height=5)

plot(x - 1, dx, col='blue', type='l', xlab='X', ylab='Density', xlim=c(-1, 9), xaxt='n')
axis(1, at = seq(-1, 9, by=1), las=1)
```

## PDF of Sum of Exponential Variables

```r
generate_data = function (n, n_samples) {
    my = matrix(rep(NA, n * n_samples), ncol=n)

    for (i in 1:n) {
        my[, i] = rnorm(n_samples, 0, 1) + rexp(n_samples, 1) - 1
    }

    dfy = data.frame(my)

    y_hat = rowMeans(dfy)
    mean(y_hat)

    return(y_hat)
}
```

## Data Generation & PLots

```r
n = 50
n_samples = 10000
hist(generate_data(n, n_samples), breaks=50, probability=T, col='gray10', density=10, xlab=expression(bar(Y)), main='')
```

```r
n=5
hist(generate_data(n, n_samples), breaks=50, probability=T, col='gray10', density=10, xlab=expression(bar(Y)), main='')
```

## Remarks
When $n$ is small (5 for example), the distribution of the sum is skewed to the right. When $n$ is large (50 for example), the distribution of the sum is approximately normal.
